# Pizza Preferences Survey Results

We asked about favorite types of pizza, and these are our results:

|Cheese|Pepperoni|Hawaiian|Other|
|-|-|-|-|
|9|7|6|3|

```python
import matplotlib.pyplot as plt
%matplotlib inline
labels = ['Pepperoni', 'Hawaiian', 'Cheese', 'Other']
sizes = [7, 6, 9, 3]
plt.title('Pizza Preferences')
plt.bar(labels, sizes)
plt.show()
```

```python
labels = ['Pepperoni', 'Hawaiian', 'Cheese', 'Other']
sizes = [7, 6, 9, 3]
plt.title('Pizza Preferences Pie')
plt.pie(sizes, labels=labels)
plt.axis('equal')
plt.show()
```

## Conclusion

We found that cheese pizza was the most favorite, then pepperoni then Hawaiian.

There were three people that liked types of pizza that were not in our tally chart, maybe next time we should have more options.
